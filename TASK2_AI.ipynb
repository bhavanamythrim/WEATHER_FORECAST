```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
#load dataset
df=pd.read_csv('/content/drive/MyDrive/spam.csv', encoding='latin1')
```

```python
df.head(10)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
x=df['v2']
y=df['v1']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
vectorizer=TfidfVectorizer()
x_train_feature=vectorizer.fit_transform(x_train)
x_test_feature=vectorizer.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train_feature,y_train)
```

```python
y_pred = model.predict(x_test_feature)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
sample_message=x_test_feature[:10]
```

```python
probabilities=model.predict_proba(sample_message)
```

```python
probabilities
```

```python
print("Categories:", y.unique())

for i, prob in enumerate(probabilities):
    sigmoid_output = prob[1]
    prediction_value = np.round(sigmoid_output)
    label = "Spam" if prediction_value == 1 else "Not Spam"

    print(f"Message {i+1}: Sigmoid Output = {sigmoid_output:.4f} ==> Prediction: {label}")
```
